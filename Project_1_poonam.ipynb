```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
CTE_student_data = './Clean_and_Combine.csv'
CTE_student_df = pd.read_csv(CTE_student_data)

CTE_student_df
```

```python
# Calculate the sum for cte students
cte_students_2015 = CTE_student_df["CTE_STUDENTS_2015"].sum()
cte_students_2016 = CTE_student_df["CTE_STUDENTS_2016"].sum()
cte_students_2017 = CTE_student_df["CTE_STUDENTS_2017"].sum()
cte_students_2018 = CTE_student_df["CTE_STUDENTS_2018"].sum()
all_students_2015 = CTE_student_df["ALL_STUDENTS_2015"].sum()
all_students_2016 = CTE_student_df["ALL_STUDENTS_2016"].sum()
all_students_2017 = CTE_student_df["ALL_STUDENTS_2017"].sum()
all_students_2018 = CTE_student_df["ALL_STUDENTS_2018"].sum()
```

```python
N = 4
CTE_total = [cte_students_2015,cte_students_2016,cte_students_2017,cte_students_2018]
CTE_arr = tuple(CTE_total)


All_total = [all_students_2015,all_students_2016,all_students_2017,all_students_2018]
All_arr = tuple(All_total)
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, CTE_arr, width, color='b', yerr=None)
rects2 = ax.bar(ind + width, All_arr, width, color='y', yerr=None)
# add some text for labels, title and axes ticks
ax.set_ylabel('Counts')
ax.set_title('Count of all students and CTE students')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('2015', '2016', '2017','2018'))

ax.legend((rects1[0], rects2[0]), ('CTE Students', 'All Students'))

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()


```
