```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
CTE_student_data = './Clean_and_Combine.csv'
CTE_student_df = pd.read_csv(CTE_student_data)

CTE_student_df.head()
```

```python
# Calculate the sum for cte students
cte_students_2015 = CTE_student_df["CTE_STUDENTS_2015"].sum()
cte_students_2016 = CTE_student_df["CTE_STUDENTS_2016"].sum()
cte_students_2017 = CTE_student_df["CTE_STUDENTS_2017"].sum()
cte_students_2018 = CTE_student_df["CTE_STUDENTS_2018"].sum()
all_students_2015 = CTE_student_df["ALL_STUDENTS_2015"].sum()
all_students_2016 = CTE_student_df["ALL_STUDENTS_2016"].sum()
all_students_2017 = CTE_student_df["ALL_STUDENTS_2017"].sum()
all_students_2018 = CTE_student_df["ALL_STUDENTS_2018"].sum()
```

```python
# Create a group based on the values in the 'maker' column
district_group = CTE_student_df.groupby('District')

district_group.head()
```

```python
Percent_enrolled_2015 = (cte_students_2015/all_students_2015)*100
Percent_enrolled_2016 = (cte_students_2016/all_students_2016)*100
Percent_enrolled_2017 = (cte_students_2017/all_students_2017)*100
Percent_enrolled_2018 = (cte_students_2018/all_students_2018)*100
```

```python
Percent_students = [Percent_enrolled_2015,Percent_enrolled_2016,Percent_enrolled_2017,Percent_enrolled_2018]
Percent_students_arr = tuple(Percent_students)
```

```python
x_axis = [2015,2016,2017,2018]
```

```python
# Count how many total students are in CTE in each district
sum_cte_district = district_group['CTE_STUDENTS_2018'].sum()

sum_cte_district
```

```python
plt.plot(x_axis, Percent_students_arr)
# Give our graph axis labels
plt.xlabel("Years")
plt.ylabel("Percentage of CTE student enrollment")

plt.show()
```
